```python
import pandas as pd
ff = pd.read_csv('results.csv')
ff
```

```python
#data of the matches played in Scotland and the home team lost is:
ff[(ff.tournament == 'FIFA World Cup qualification') & (ff.country == 'Scotland') & (ff.home_score==0) & (ff.away_score!=0)]
```

```python
#number of matches played in Nehru Cup where home team won by more than 2 goals
ff[ (ff.tournament == 'Nehru Cup') & (ff.home_team == 'India') & (ff.home_score - ff.away_score >2)]
```

```python
#statistics of away score of all the matches in EURO CUP
ff.agg({'home_score': ['min', 'max', 'mean', 'median']})
```

```python
#matches played in scotland grouped by home_team and filtered on basis of home_score sorted in ascending order
ff[ff.country == 'Soviet Union'].groupby('home_team').filter(lambda g: len(g) > 2).groupby('home_score').size().sort_values(ascending=False)
```

```python
#A dictionary which displays the year and the venue of FIFA 
ff['year'] = pd.DatetimeIndex(ff['date']).year
fifa = ff[ff['tournament']=='FIFA World Cup']
dates = fifa.date.tolist()
years=[]
for date in dates:
    year=date.split('-')[0]
    if year not in years:
        years.append(year)
dict_country={}
for i in range(len(years)):
    for j in range(len(fifa[fifa['year']==eval(years[i])].country.unique())):
        if years[i] in dict_country.keys():
            dict_country[years[i]].append(fifa[fifa['year']==eval(years[i])].country.unique()[j])
        else:
            dict_country[years[i]]=[fifa[fifa['year']==eval(years[i])].country.unique()[j]]
dict_country
```
