```python
import time
grpa=[['INDIA',0],['AUSTRALIA',0],['NEWZEALAND',0],['KENYA',0]]
grpb=[['WINDIES',0],['BANGLADESH',0],['ENGLAND',0],['SCOTLAND',0]]
grpc=[['IRELAND',0],['AFGHANISHTAN',0],['PAKISTAN',0],['SRILANKA',0]]
grpd=[['ZIMBABWE',0],['CANADA',0],['UAE',0],['NAMIBIA',0]]

def play(listp):
    for i in range(0,len(listp)):
        for j in range(i+1,len(listp)):
            if(listp[i]==listp[j]):
                pass
            else:
                print("\nThe match between", listp[i][0],"and ", listp[j][0])
                point=eval(input("Enter 2 for win,1 for draw :- "))
                if point==2:
                    wd=input("Winning team : ")
                    if wd.lower()==listp[j][0].lower():
                        listp[j][1]=listp[j][1]+2
                    elif wd.lower()==listp[i][0].lower():
                        listp[i][1]=listp[i][1]+2
                elif point==1:
                    listp[i][1]=listp[i][1]+1
                    listp[j][1]=listp[j][1]+1

def Sort(unsorted_list): 
    unsorted_list.sort(key = lambda x: x[1],reverse=True) 
    return unsorted_list 

print("The groups along with their teams")
print("Group A(0) : INDIA, AUSTRALIA, NEWZEALAND, KENYA")
time.sleep(2)
print("Group B(1) : WINDIES, BANGLADESH, ENGLAND, SCOTLAND")
time.sleep(2)
print("Group C(2) : IRELAND, AFGHANISHTAN, PAKISTAN, SRILANKA")
time.sleep(2)
print("Group D(3) : ZIMBABWE, CANADA, UAE, NAMIBIA")
print()
print()

play(grpa)
play(grpb)
play(grpc)
play(grpd)

time.sleep(4)
print("\nFinal Team scores after the matches. Top two teams will be selected for the next round")
sorted_grpa = Sort(grpa)
print(sorted_grpa)
sorted_grpb = Sort(grpb)
print(sorted_grpb)
sorted_grpc = Sort(grpc)
print(sorted_grpc)
sorted_grpd = Sort(grpd)
print(sorted_grpd)
print()
print()

time.sleep(2)
print("Teams for the next round")
grpe=[[sorted_grpa[0][0],0],[sorted_grpa[1][0],0],[sorted_grpb[0][0],0],[sorted_grpb[1][0],0]]
grpf=[[sorted_grpc[0][0],0],[sorted_grpc[1][0],0],[sorted_grpd[0][0],0],[sorted_grpd[1][0],0]]
print(grpe)
print(grpf)

play(grpe)
play(grpf)
print()
print()

time.sleep(4)
print("\nFinal Team scores after the matches. Top two teams will be selected for the final round")
sorted_grpe = Sort(grpe)
print(sorted_grpe)
sorted_grpf = Sort(grpf)
print(sorted_grpf)
print()

time.sleep(2)
print("Teams for the final round")
grpg=[[sorted_grpe[0][0],0],[sorted_grpe[1][0],0],[sorted_grpf[0][0],0],[sorted_grpf[1][0],0]]
print(grpg)  

play(grpg)

time.sleep(4)
print("\n\nFinal standing after all the matches")
sorted_grpg = Sort(grpg)
print(sorted_grpg)
print()

time.sleep(4)
print ("The winner is :")
print (sorted_grpg[0][0])
```
